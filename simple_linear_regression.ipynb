```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/data/placement (1).csv")
df
```

```python
# find the numner of rows and column
df.shape
```

```python
# find the information of data
df.info()
```

```python
# find the null value in the dataset
df.isnull().sum()
```

```python
# find the independent feature and dependent feature
x=df.iloc[:,0].values
y=df.iloc[:,1].values
x
```

```python
# split the data into ttain and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train.reshape(-1,1),y_train)
```

```python
# coefficent
m = model.coef_
print(m[0])
```

```python
# intercept
b = model.intercept_
print(b)
```

```python
cgpa = 9.5
package = m*cgpa + b
print(package)
```

```python
y_pred = model.predict(x_test.reshape(-1,1))
y_pred, y_test
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["cgpa"], df["package"])
plt.plot(x_train, m*x_train + b, color="red")
plt.xlabel("cgpa")
plt.ylabel("package")
plt.show()
```
